```python
import pandas as pd
df=pd.read_csv("D:\\Downloads\\studentmarks1.csv")
print(df)
```

```python
df['Percent']=round((df['Total']/600)*100,2)
print(df)
```

```python
df.loc[(df["Percent"]>=60) & (df["Percent"]<=80),['grade']]='first'
print(df)
```
